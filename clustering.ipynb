```python
import pandas as pd
import numpy as np
from sklearn.cluster import OPTICS
from sklearn.cluster import SpectralClustering
from sklearn.cluster import KMeans

import matplotlib.pyplot as plt


## Loading Data
sales = pd.read_csv("data/sales_train.csv")
categories = pd.read_csv("data/item_categories.csv")
test_set = pd.read_csv("data/test.csv")
items = pd.read_csv(f"data/items.csv")
```

```python
sales = sales.merge(items[["item_id", "item_category_id"]], how="left", on="item_id")
```

```python
cluster_df = sales[["item_category_id", "item_cnt_day", "date_block_num"]].groupby(by=["date_block_num","item_category_id"], as_index=False).sum()
```

```python
optics = OPTICS()
spectral = SpectralClustering()
kmeans = KMeans()
```

```python
plt.scatter(cluster_df["item_category_id"], cluster_df["item_cnt_day"], c=cluster_df["date_block_num"])
plt.show()
```

```python
plt.scatter(cluster_df["item_category_id"], cluster_df["item_cnt_day"], c=optics.fit_predict(X=cluster_df))
plt.show()
```
